For the function get_interest_points, xs and ys are arrays holding the coordinates of the interest points in the image. For this one, I first used gaussian filter on the image. Then I used a horizontal and vertical sobel filter seperately on the image for the spatial derivative calculation part. Next, I convolved gaussian filter with Sx^2, Sx*Sy, Sy^2 to get Ix2, Ixy and Iy2 for the structure tensor setup. For corner response calculation, I used 2 for loops and calculated the determinant and trace of the matrix M, and I chose a threshold of 0.1. For this threshold, as it gets larger, the result tends to get worse, and if change it to 0.05, then it would generate even better result, as more interest points would be chosen. Lastly, I performed the Non-Max suppression procedure with the use of the funciton filters.perk_local_max, and got the result. Next, for the function get_features, I firstly used gaussian filter on the image. Then in order to get the magnitude and orientation matrix for the image, I used horizontal and vertical sobel filters to get dy and dx, and used them to calculated the magnitude and orientation, and I converted all negative values in orientation to positive, and this increased the accuracy result on the Mount_Rushmore by 20~30%. Then I traversed the coordinates, and for each, I draw a 16x16 window, and find the corresponding magnitude and orientation of the window. Within the 16x16 window, I find the 4x4 window and its corresponding magnitude and orientation. Then, within in each 4x4 window, for 8 directions, each with 45 degrees, I calculated their orientation, combined them and put the result in the features array. Lastly, I normalized the result and returned it. For the function match_features, by using the for loop to traverse each features, I calculated the distance between the feature in first image and every other features in the second image, and they are stored in an array. Then, I found the smallest and second smallest distances, and used them to calculate the Nearest Neighbor Distance Ratio. And for this one, I chose the threshold to be 0.9 and it has the best result amongst the images. If the Nearest Neighbor Distance Raio is smaller than this raito, then I would say there's a match, and the confidence score is calculated as 1/NNDR.
